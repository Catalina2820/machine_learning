## In Synthetic data

````python
import numpy as np # Librer√≠a mateamtica
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import make_regression
````

````python
X, y = make_regression(n_samples = 500, n_features = 1, noise = 10, random_state = 1)
y = (y*0.1) **2
_, ax = plt.subplots (figsize=(5,4))
ax.scatter( X , y , s=10, c = 'b');
````

````python
svr = SVR(kernel = "rbf" )

svr.fit( X.reshape(-1,1) , y )
````

````python
# Predecir la nueva muestra. Imprimir la prediccion
prediction = svr.predict(X.reshape(-1,1))
print (prediction.shape)
````

````python
# Predict x values
# predictions = _

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , s=10, color = 'blue')
ax.scatter ( X , prediction , s=10, color = 'red')
# ax.scatter(X, y, svr.support_vectors_, s=30, edgecolors='k', fc='none')
````

## In real data

```datasets``` module info [here](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)

````python
import numpy as np
import matplotlib.pyplot as plt # plots

from sklearn.svm import SVR
from sklearn.datasets import load_diabetes
````

````python
# Cargar la base de datos Diabetes
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)
print( diabetes_X.shape , diabetes_y.shape )
````

````python
# Use only one feature
diabetes_X = diabetes_X[:,2].reshape(-1,1)
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Linear regression
svr = SVR(kernel='rbf')

# Train the model
svr.fit(diabetes_X, diabetes_y)
````

````python
# Make predictions using the testing set
diabetes_y_pred = svr.predict(diabetes_X)
print(diabetes_y_pred.shape)
````

````python
# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( diabetes_X , diabetes_y , color = 'blue')
ax.scatter( diabetes_X , diabetes_y_pred , color = 'red');
````
