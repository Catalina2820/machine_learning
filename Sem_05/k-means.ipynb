# K Means - Generated data

Isotropic dataset module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) \
Train-test splits in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) \
k-Means module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```
X, _ = make_blobs(n_samples = 400 , n_features= 2 ,random_state = 1 ) #No tenemos etiquetas, no hay 'y'

# Create train (80%) and test (20%) sets and print all shapes
X_train, X_test = train_test_split(X, test_size = 0.2, random_state= 1)

_, axes = plt.subplots(1,2, figsize=(8,4))
axes[0].scatter( X[:,0] , X[:,1] , edgecolors='k', c='g')
axes[1].scatter( X_train[:,0] , X_train[:,1] , edgecolors='k', c='b')
axes[1].scatter( X_test[:,0] , X_test[:,1] , edgecolors='k', c='r');
```

```
kmeans = KMeans(n_clusters=3, random_state= 1) #al ver la grfica elijo la cantidad de clusters

# Train Kmeans
kmeans.fit(X_train)
```

```
# Plot centroids
centroids = kmeans.cluster_centers_
print(centroids)
```

```
_, ax = plt.subplots (figsize = (5,4))
colors=['blue','cyan','magenta']

# plot data and found centroids
ax.scatter(X_test[:,0], X_test[:,1], c='b', edgecolors='k')
ax.scatter(centroids[:,0], centroids[:,1], s=700 , marker = '*' , edgecolors='k', c='y');
```

```
# Create predictions
predictions = kmeans.predict(X_test)
```

```
# Plot decision boundaries
_, ax = plt.subplots(figsize = (5,4))
ax.scatter(X_test[:,0], X_test[:,1], c= predictions, edgecolors='k', cmap ='Paired');
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

```
data = pd.read_csv('/content/drive/MyDrive/Machine learning/machine_learning/Sem_03/fruits_data_ext.csv')
data.head()
```

```
X = data[['width','color_score']].values
# Y = _
```

```
# Split train / test data
X_train, X_test = train_test_split(X, test_size = 0.2, random_state= 1)
```

```
# Train Kmeans
kmeans = KMeans(n_clusters= 4, random_state=1)
kmeans.fit(X_train)
```

```
centroids = kmeans.cluster_centers_
print(centroids)
```

```
_, ax = plt.subplots (figsize = (5,4))
colores=['blue','cyan','magenta']

ax.scatter(X_test[:,0], X_test[:,1], c='b', edgecolors='k')
ax.scatter(centroids[:,0], centroids[:,1], s=400, marker= 'x', edgecolors='k', c='r');
```

```
# Predict values
predictions = kmeans.predict(X_test)
```

```
# Plot decision boundaries
_, ax = plt.subplots(figsize = (5,4))
ax.scatter(X_test[:,0], X_test[:,1], c= predictions, edgecolors='k', cmap ='Paired');
```
