````python
import numpy as np # Numerical library
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression
````

````python
def f(x) : # Create a function that samples from linear function
  y = (0.0918 * x) + 1.2859 + 0.15*np.random.randn (x.shape[0]) # Linear function, el 0.15 es el ruido
  return y

# create training samples
x = np.arange (0, 20, 0.25); print (x.shape) #da 80 puntos
# Generate samples
y = f(x) ; print (y.shape)

# Scatter plot
_, ax = plt.subplots(figsize=(5,4))
ax.scatter(x, y, color = 'blue')
plt.show()
````

````python
# Create an instance
linear_regression = LinearRegression()

# Train the model
linear_regression.fit( x.reshape(-1,1) , y ) # (# Samples, # Features)
#se debe hacer el reshape dde x porque tenemos una sola características y sklearn recibe un  arreglo 2D
````

````python
# Print estimated values
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# Create a new sample estimate prediction
new_sample = np.array([5])
print (f(new_sample))

prediction = linear_regression.predict( new_sample.reshape(-1,1) )
print (prediction)
````

````python
# Predict x values
predictions = linear_regression.predict( x.reshape(-1,1) )

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( x , y , color = 'blue')
ax.plot ( x , predictions , color = 'red')
````

## In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
````

````python
X, y = make_regression(n_samples = 100 , n_features = 1, noise = 10, random_state = 5 )

_, ax = plt.subplots (figsize=( 5 , 4 ))
ax.scatter( X , y , s =10, c = 'b')
print (X.shape, y.shape)
````

````python
linear_regression = LinearRegression()
linear_regression.fit( X.reshape(-1,1) , y )
````

````python
# Imprimir los valores estimados para compararlos respecto a los originales
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# Nueva muestra. Imprime el valor correspondiente de y
# new_sample = np.array([1])

# Predecir la nueva muestra. Imprimir la prediccion
predictions = linear_regression.predict( X.reshape(-1,1) )
print (predictions.shape)
````

````python
_, ax = plt.subplots(figsize=(5,4))
ax.scatter(X, y, s=10, c='b')
ax.plot(X, predictions, c='g', lw=1);
````

## In real data

```datasets``` module info [here](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.datasets)

````python
import numpy as np
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression # scikit-learn: libreria de IA para python
from sklearn.datasets import load_diabetes
````

````python
# Cargar la base de datos Diabetes
diabetes_X, diabetes_y = load_diabetes(return_X_y=True)
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Use only one feature
diabetes_X = diabetes_X[:,2].reshape(-1,1)
diabetes_y = diabetes_y
print(diabetes_X.shape, diabetes_y.shape)
````

````python
# Linear regression
linear_regression = LinearRegression()

# Train the model
linear_regression.fit( diabetes_X , diabetes_y )
````

````python
# Make predictions using the testing set
diabetes_y_pred = linear_regression.predict(diabetes_X)


print ("w = " + str( linear_regression.coef_ ) + ", b = " + str( linear_regression.intercept_ ))
````

````python
# Plot outputs
_, ax = plt.subplots (figsize=( 5 , 4 ))
ax.scatter( diabetes_X, diabetes_y, s=10, color = "blue")
ax.plot(diabetes_X, diabetes_y_pred, color = "red");
````
